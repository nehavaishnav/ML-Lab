```python
import pandas as pd

data = pd.read_csv("training.csv")

X = data.iloc[:, 0:-1].values
y = data.iloc[:, -1].values

for i in range(len(y)):
    if y[i].strip().lower() == 'yes':
        S = X[i].copy()
        break

G = [['?' for _ in range(len(S))]]

for i in range(len(X)):
    if y[i].strip().lower() == 'yes':
        for j in range(len(S)):
            if X[i][j] != S[j]:
                S[j] = '?'

        G = [g for g in G if all(
            g[k] == '?' or g[k] == S[k] for k in range(len(S))
        )]

    else:
        new_G = []
        for g in G:
            for j in range(len(S)):
                if g[j] == '?' and S[j] != X[i][j]:
                    temp = g.copy()
                    temp[j] = S[j]
                    if temp not in new_G:
                        new_G.append(temp)

        if new_G:
            G = new_G

print("Find S:", S)
print("Find G:", G)
```
